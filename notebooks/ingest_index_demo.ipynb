## Person A: Load and Inspect Logical View

Choose either Project Gutenberg (`pg_catalog.csv`) or the Kaggle 7k books dataset.

```python
# Set dataset_root to the folder where your CSV lives
dataset_root = 'data/raw'

pg_path = find_pg_catalog(dataset_root)
kaggle_path = find_books_csv(dataset_root)

print('pg_catalog.csv:', pg_path)
print('kaggle books csv:', kaggle_path)
```

```python
# Load records from whichever dataset you have available
records = []
if pg_path:
    records = load_pg_catalog(pg_path, filter_english=True)
    dataset_name = 'pg_catalog'
elif kaggle_path:
    records = load_books_csv(kaggle_path)
    dataset_name = 'kaggle_books'
else:
    dataset_name = None

print('dataset:', dataset_name)
print('records:', len(records))
```

```python
# Show a few logical-view records
records[:3]
```

## Person B: Build Index + Search

Build the inverted index and run a TF-IDF search query.

```python
index = InvertedIndex.build(records)
print('indexed docs:', len(index.doc_term_freqs))
print('vocab size:', len(index.postings))
```

```python
query = 'mystery detective london'
results = search(index, query, top_k=10)
[(r.book_id, r.display.get('title', ''), r.display.get('authors', ''), r.score) for r in results]
```
